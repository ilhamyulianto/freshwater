```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# BEFORE

```python
df = pd.read_excel('../data/freshwater_datamaster.xlsx', sheet_name='before')
df = df[['SHIP-ROUTE','DISTANCE', 'FRESH_WATER']]
```

```python
df['FRESH_WATER'] = pd.to_numeric(df['FRESH_WATER'], errors='coerce')
df = df.dropna(subset=['FRESH_WATER'])


plt.figure(figsize=(10, 6))
sns.kdeplot(df['FRESH_WATER'], fill=True, color='skyblue')
plt.title('Density Plot of Fresh Water Consumption')
plt.xlabel('Fresh Water')
plt.ylabel('Density')
plt.grid(True)
plt.tight_layout()
plt.show()


bins = [0, 2500, 5000, 7500, 10000, 12500, 15000, 17500, 20000]
labels = ['0-2500', '2501-5000', '5001-7500', '7501-10000', '10001-12500','12501-15000', '15001-17500', '17501-20000']
df['RANGE'] = pd.cut(df['FRESH_WATER'], bins=bins, labels=labels, include_lowest=True)
```

```python

range_counts = df['RANGE'].value_counts().sort_index()


plt.figure(figsize=(10, 8))
sns.barplot(x=range_counts.index, y=range_counts.values, palette='Blues_d')
plt.title('Fresh Water Consumption Range Distribution')
plt.xlabel('Range')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```
